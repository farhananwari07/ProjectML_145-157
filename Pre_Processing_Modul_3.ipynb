# Pre Processing

```python
# TULIS KODE ANDA DISINI
from keras.preprocessing.image import ImageDataGenerator
```

```python
augs_gen = ImageDataGenerator(
    rescale=1./255,        
    horizontal_flip=True,
    height_shift_range=.2,
    vertical_flip = True
)  
```

```python
train_dir = '/content/drive/MyDrive/Semester 7//Machine Learning/File Praktikum/Modul 3/model_covid/train'
train_gen = augs_gen.flow_from_directory(
    train_dir,
    target_size = (120,120),
    batch_size=32,
    color_mode='grayscale',
    class_mode = 'binary'
)
```

```python
val_dir ='/content/drive/MyDrive/Semester 7//Machine Learning/File Praktikum/Modul 3/model_covid/validation'
val_gen = augs_gen.flow_from_directory(
    val_dir,
    target_size=(120,120),
    batch_size=32,
    color_mode='grayscale',
    class_mode='binary'
)
```

```python
test_dir = '/content/drive/MyDrive/Semester 7//Machine Learning/File Praktikum/Modul 3/model_covid/test'
test_gen = augs_gen.flow_from_directory(
    test_dir,
    target_size=(120,120),
    batch_size=32,
    color_mode='grayscale',
    class_mode='binary'
)
```

```python
import numpy as np

x_train=np.concatenate([train_gen.next()[0] for i in range(train_gen.__len__())])
y_train=np.concatenate([train_gen.next()[1] for i in range(train_gen.__len__())])
print(x_train.shape)
print(y_train.shape)
```

```python
x_val=np.concatenate([val_gen.next()[0] for i in range(val_gen.__len__())])
y_val=np.concatenate([val_gen.next()[1] for i in range(val_gen.__len__())])
print(x_val.shape)
print(y_val.shape)
```

```python
x_test=np.concatenate([test_gen.next()[0] for i in range(test_gen.__len__())])
y_test=np.concatenate([test_gen.next()[1] for i in range(test_gen.__len__())])
print(x_test.shape)
print(y_test.shape)
```

```python
np.save("xtrain64", x_train)
np.save("xval64", x_val)
np.save("xtest64", x_test)
np.save('ytrain64', y_train)
np.save('yval64', y_val)
np.save('ytest64', y_test)
```
